```python
import requests
from io import StringIO

url_exp = "https://raw.githubusercontent.com/Honza-R/PL-REX/main/009-CDK2/experimental_dG.txt"
response_exp = requests.get(url_exp)
exp_txt = response_exp.text
```

```python
rows_exp = []
for line in exp_txt.splitlines():
    line = line.strip()
    if not line or line.startswith("#"):
        continue
    parts = line.split()
    # expected: complex_id  DGexp
    rows_exp.append((parts[0], float(parts[1])))

df_exp = pd.DataFrame(rows_exp, columns=["complex_id", "DGexp"])
df_exp
```

```python
url_sqm = "https://raw.githubusercontent.com/Honza-R/PL-REX/main/009-CDK2/score_SQM2.20.txt"
response_sqm = requests.get(url_sqm)
sqm_txt = response_sqm.text
```

```python
rows_sqm = []
for line in sqm_txt.splitlines():
    line = line.strip()
    if not line or line.startswith("#"):
        continue
    parts = line.split()
    rows_sqm.append((parts[0], float(parts[1])))

df_sqm = pd.DataFrame(rows_sqm, columns=["complex_id", "SQM2_score"])
df_sqm.head()
```

```python
df_compare = df_exp.merge(df_sqm, on="complex_id", how="inner")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter

# --- Clean data (this is the usual reason R or R^2 becomes NaN) ---
tmp = df_compare[["DG_tot", "DGexp", "complex_id"]].copy()

# Convert to numeric (turns bad strings like "NA", "" into NaN)
tmp["DG_tot"] = pd.to_numeric(tmp["DG_tot"], errors="coerce")
tmp["DGexp"]  = pd.to_numeric(tmp["DGexp"], errors="coerce")

# Remove inf and NaN
tmp = tmp.replace([np.inf, -np.inf], np.nan).dropna(subset=["DG_tot", "DGexp"])

# Extract arrays after cleaning
x = tmp["DG_tot"].to_numpy()
y = tmp["DGexp"].to_numpy()
labels = tmp["complex_id"].to_numpy()

# --- Compute correlation safely ---
if len(x) < 2 or np.std(x) == 0 or np.std(y) == 0:
    R = np.nan
    R2 = np.nan
else:
    R = np.corrcoef(x, y)[0, 1]
    R2 = R**2

# Optional: print what got used
print(f"Rows used for correlation/plot: {len(tmp)}")
print(f"R = {R}, R2 = {R2}")

# --- Plot ---
plt.figure(figsize=(20, 20))
plt.scatter(x, y, s=30)

dx, dy = 0.15, 0.15
for xi, yi, label in zip(x, y, labels):
    plt.text(xi + dx, yi + dy, str(label), fontsize=20)

plt.xlabel("Computed Delta G (Binding, kcal/mol)", fontsize=20)
plt.ylabel("DGexp", fontsize=20)

# Avoid formatting NaN with :.2f (it will still print 'nan', but this is safer)
title_R = "nan" if np.isnan(R) else f"{R:.2f}"
title_R2 = "nan" if np.isnan(R2) else f"{R2:.2f}"
plt.title(f"09-CDK2, R = {title_R}   R$^2$ = {title_R2}", fontsize=20)

# xticks need finite min/max
if len(x) > 0 and np.isfinite(x).all():
    plt.xticks(np.arange(np.min(x), np.max(x) + 5, 5), fontsize=20)
plt.yticks(fontsize=20)

plt.tight_layout()
plt.savefig("Delta G Binding_vs_Experimental.jpg", dpi=100)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats

# =========================
# 1) Clean / prepare data
# =========================
tmp = df_compare[["DG_tot", "DGexp", "complex_id"]].copy()

# Convert to numeric (bad strings -> NaN)
tmp["DG_tot"] = pd.to_numeric(tmp["DG_tot"], errors="coerce")
tmp["DGexp"]  = pd.to_numeric(tmp["DGexp"],  errors="coerce")

# Remove inf and NaN
tmp = tmp.replace([np.inf, -np.inf], np.nan).dropna(subset=["DG_tot", "DGexp"])

# Extract arrays after cleaning
x = tmp["DG_tot"].to_numpy()
y = tmp["DGexp"].to_numpy()
labels = tmp["complex_id"].to_numpy()

print(f"Rows used for correlation/plot: {len(tmp)}")

# =========================
# 2) Correlation + p-value
# =========================
if len(x) < 2 or np.std(x) == 0 or np.std(y) == 0:
    R = np.nan
    R2 = np.nan
    p_value = np.nan
else:
    R, p_value = stats.pearsonr(x, y)
    R2 = R**2

print(f"R = {R}, R2 = {R2}, p-value = {p_value}")

# ==========================================
# 3) Bootstrap 95% CI for Pearson correlation
# ==========================================
n_boot = 10000
rng = np.random.default_rng(0)  # set seed for reproducibility
boot_R = []

if len(x) >= 3 and np.std(x) != 0 and np.std(y) != 0:
    for _ in range(n_boot):
        idx = rng.integers(0, len(x), len(x))  # resample indices with replacement
        xb = x[idx]
        yb = y[idx]

        # Guard against degenerate bootstrap samples
        if np.std(xb) == 0 or np.std(yb) == 0:
            continue

        r_boot = np.corrcoef(xb, yb)[0, 1]
        if np.isfinite(r_boot):
            boot_R.append(r_boot)

    if len(boot_R) > 20:  # arbitrary minimum so CI isn't nonsense
        ci_low, ci_high = np.percentile(boot_R, [2.5, 97.5])
    else:
        ci_low, ci_high = np.nan, np.nan
else:
    ci_low, ci_high = np.nan, np.nan

print(f"Bootstrap 95% CI for R: [{ci_low}, {ci_high}] (n={len(boot_R)} boot samples used)")

# =========================
# 4) Regression line + stats
# =========================
if len(x) >= 2 and np.std(x) != 0:
    slope, intercept = np.polyfit(x, y, 1)
else:
    slope, intercept = np.nan, np.nan

# Prediction metrics (optional but recommended)
rmse = np.sqrt(np.mean((x - y) ** 2)) if len(x) > 0 else np.nan
mae  = np.mean(np.abs(x - y)) if len(x) > 0 else np.nan

print(f"slope = {slope}, intercept = {intercept}")
print(f"RMSE = {rmse}, MAE = {mae}")

# =========================
# 5) Plot
# =========================
plt.figure(figsize=(20, 20))
plt.scatter(x, y, s=30)

# Label points
dx, dy = 0.15, 0.15
for xi, yi, label in zip(x, y, labels):
    plt.text(xi + dx, yi + dy, str(label), fontsize=20)

plt.xlabel("Computed ΔG (Binding, kcal/mol)", fontsize=20)
plt.ylabel("Experimental ΔG (kcal/mol)", fontsize=20)

# Best-fit regression line
if np.isfinite(slope) and np.isfinite(intercept):
    x_sorted = np.sort(x)
    y_fit = slope * x_sorted + intercept
    plt.plot(x_sorted, y_fit, linewidth=3, label="Best-fit line")

# y = x reference line
if len(x) > 0:
    min_val = min(np.min(x), np.min(y))
    max_val = max(np.max(x), np.max(y))
    plt.plot([min_val, max_val], [min_val, max_val], linestyle="--", linewidth=2, label="y = x")

# Title formatting (safe for NaNs)
R_str   = "nan" if np.isnan(R) else f"{R:.2f}"
R2_str  = "nan" if np.isnan(R2) else f"{R2:.2f}"
p_str   = "nan" if np.isnan(p_value) else f"{p_value:.3g}"
ci_str  = "nan" if (np.isnan(ci_low) or np.isnan(ci_high)) else f"[{ci_low:.2f}, {ci_high:.2f}]"
rmse_str = "nan" if np.isnan(rmse) else f"{rmse:.2f}"
mae_str  = "nan" if np.isnan(mae) else f"{mae:.2f}"

plt.title(
    f"09-CDK2\n"
    f"Pearson R = {R_str} (95% CI {ci_str}), R$^2$ = {R2_str}, p = {p_str}\n"
    f"Fit: y = {slope:.2f}x + {intercept:.2f}   |   RMSE = {rmse_str}, MAE = {mae_str}",
    fontsize=20
)

# Ticks (only if x is finite)
if len(x) > 0 and np.isfinite(x).all():
    step = 5
    xmin, xmax = np.min(x), np.max(x)
    plt.xticks(np.arange(np.floor(xmin / step) * step, np.ceil(xmax / step) * step + step, step), fontsize=20)
plt.yticks(fontsize=20)

plt.legend(fontsize=18)
plt.tight_layout()

plt.savefig("DeltaG_Binding_vs_Experimental_with_stats.jpg", dpi=150)
plt.show()
```
